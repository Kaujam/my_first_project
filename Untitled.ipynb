```python
import requests
resp = requests.get("https://docs.scipy.org/doc/scipy/release/1.14.1-notes.html")
print(resp.text)
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(resp.text,'html.parser')
print(soup.prettify())
```

```python
items = soup.find_all('li')[110:153]

# 定义版本信息
version = "1.14.1"

# 创建一个空列表来保存所有标签及其对应的版本信息的字典
labels_with_versions = []

# 遍历<li>标签，仅提取标签并保存到字典中，同时添加版本信息
for item in items:
    text = item.get_text()
    # 假设标签总是出现在冒号前且为大写单词
    label = next((part.strip() for part in text.split(':') if part.strip().isupper()), 'UNKNOWN')
    labels_with_versions.append({'label': label, 'version': version})

# 打印所有标签及其对应的版本信息
print(labels_with_versions)
```

```python
import matplotlib.pyplot as plt
from collections import Counter

# 统计每种标签的数量
label_counts = Counter(item['label'] for item in labels_with_versions)

# 提取标签和对应的数量用于绘图
labels, counts = zip(*label_counts.items())

# 创建柱状图
plt.figure(figsize=(10, 6))
plt.bar(labels, counts, color='skyblue')

# 添加标题和标签
plt.title('Label Frequency Distribution')
plt.xlabel('Labels')
plt.ylabel('Frequency')
plt.xticks(rotation=45, ha='right')  # 旋转 x 轴标签以便阅读

# 显示图表
plt.tight_layout()
plt.show()

# 打印每个标签的数量
for label, count in label_counts.items():
    print(f'Label="{label}", Count={count}')
```
